```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Shoogloo(data_shoogloo):
     return data_shoogloo.columns.tolist()

def Relevent_Cols(data_shoogloo):
          return data_shoogloo.filter(items=['conversion_id', 
                                        'campaign_name',
                                        'p1',
                                        'status', 
                                        'created',
                                        'sale_amount', 
                                        'currency',
                                        'payout']
                                 )   

data_shoogloo = pd.read_csv(r"C:\Users\mk744\Downloads\Shoogloo Data Dec-23 to Jan-24.csv")
shoogloo = Shoogloo(data_shoogloo) 
shoogloo
```

```python
Rel_Cols = Relevent_Cols(data_shoogloo)
Rel_Cols
```

```python
def Add_Cols(Rel_Cols):
    # Add new columns with specified values
    Rel_Cols['Partner'] = 'shoogloo'
    Rel_Cols['Inventory'] = 'hyyzo'
    return Rel_Cols

# Apply the function and store the result
Add_new_cols = Add_Cols(Rel_Cols)
Add_new_cols.head()
```

```python
# Adjust date time
import warnings
warnings.filterwarnings('ignore')
def Date_time(Add_new_cols):
     Add_new_cols['Date'] = pd.to_datetime(Add_new_cols['created'], errors = 'coerce').dt.date
     Add_new_cols.drop(columns=['created'], inplace=True)
     return Add_new_cols

date_time = Date_time(Add_new_cols)
date_time
```

```python
# Rename Columns
def Rename_Cols(date_time):
      date_time.rename(columns={'conversion_id':'Id',
                                      "campaign_name":"Campaign_Name",
                                      "p1":"Sub_Id",
                                      "status":"Status",
                                      "sale_amount":"Amount",
                                      "currency":"Currency",
                                      "payout":"Payout"
                                   }, inplace = True
                             )
      
      return date_time
     
New_Cols = Rename_Cols(date_time)
New_Cols
```

# Amazon India

```python
Amazon_hyyzo = pd.read_csv(r"C:\Users\mk744\Downloads\Hyyzo Amazon.csv")
Amazon_hyyzo.columns.tolist()
```

```python
Amazon_hyyzo_Cols = Amazon_hyyzo.filter(items=['Unique_Id', 'Campaign_Name',
                                               "Partner",  "Amount", 
                                                "Payout", "Action_Date"])
Amazon_hyyzo_Cols['Date'] = pd.to_datetime(Amazon_hyyzo_Cols['Action_Date'],dayfirst=True, errors='coerce').dt.strftime('%Y-%m-%d')
Amazon_hyyzo_Cols.drop(columns=['Action_Date'], inplace=True)
Amazon_hyyzo_Cols
```

```python
Cue_hyyzo = pd.read_excel(r"C:\Users\mk744\Downloads\Cuelinks HYYZO.xls")
Cue_hyyzo_cols = Cue_hyyzo.query("Campaign_Name == 'Amazon India'")
```

```python
Cue_hyyzo_cols['Partner'] = 'cuelinks'
Cue_data = Cue_hyyzo_cols.filter(items=['Id', 'Campaign_Name',
                                               "Partner",  "Amount", 
                                                "Payout", "Action_Date"])
Cue_data
```

```python
# Merge the DataFrames
Merge_data = pd.merge(
        Amazon_hyyzo_Cols,
        Cue_data,
        left_on='Unique_Id',
        right_on='Id',
        how='outer',
        suffixes=("__Hyyzo", "_Partner_Hyyzo")
    )
Merge_data['User_Payout'] = Merge_data['Payout_Partner_Hyyzo'] * 0.9
Merge_data['Date'] = pd.to_datetime(Merge_data['Date'], errors = 'coerce')
Merge_data.to_excel("Amazon_hyyzo_payout.xlsx", index=False)
```

```python
Impact_cols = Impact.filter(items=['Action-Date', 'Id', 'Campaign_Name',
                                   'Event_Type', 'Status','Payout'
                                   ])
Impact_cols.head()
```

```python
Impact_cols['Date'] = pd.to_datetime(Impact_cols['Action-Date'], errors='coerce').dt.date
Impact_cols['Time'] = pd.to_datetime(Impact_cols['Action-Date'], errors='coerce').dt.time
Impact_cols.drop(columns=['Action-Date'], inplace=True)
Impact_cols.isna().sum()
```

```python
Impact_cols.groupby(['Campaign_Name', 'Event_Type']).agg(
     Total_Payout = ("Payout", lambda a: int(round(a.sum(), 2))),
     Total_Orders = ("Payout", "count"),
).reset_index().sort_values(by='Total_Payout', ascending=False)
```
