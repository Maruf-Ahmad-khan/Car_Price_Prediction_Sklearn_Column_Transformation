```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Shoogloo(data_shoogloo):
     return data_shoogloo.columns.tolist()

def Relevent_Cols(data_shoogloo):
          return data_shoogloo.filter(items=['conversion_id', 
                                        'campaign_name',
                                        'p1',
                                        'status', 
                                        'created',
                                        'sale_amount', 
                                        'currency',
                                        'payout']
                                 )   

data_shoogloo = pd.read_csv(r"C:\Users\mk744\Downloads\Shoogloo Data Dec-23 to Jan-24.csv")
shoogloo = Shoogloo(data_shoogloo) 
shoogloo
```

```python
Rel_Cols = Relevent_Cols(data_shoogloo)
Rel_Cols
```

```python
def Add_Cols(Rel_Cols):
    # Add new columns with specified values
    Rel_Cols['Partner'] = 'shoogloo'
    Rel_Cols['Inventory'] = 'hyyzo'
    return Rel_Cols

# Apply the function and store the result
Add_new_cols = Add_Cols(Rel_Cols)
Add_new_cols.head()
```

```python
# Adjust date time
import warnings
warnings.filterwarnings('ignore')
def Date_time(Add_new_cols):
     Add_new_cols['Date'] = pd.to_datetime(Add_new_cols['created'], errors = 'coerce').dt.date
     Add_new_cols.drop(columns=['created'], inplace=True)
     return Add_new_cols

date_time = Date_time(Add_new_cols)
date_time
```

```python
# Rename Columns
def Rename_Cols(date_time):
      date_time.rename(columns={'conversion_id':'Id',
                                      "campaign_name":"Campaign_Name",
                                      "p1":"Sub_Id",
                                      "status":"Status",
                                      "sale_amount":"Amount",
                                      "currency":"Currency",
                                      "payout":"Payout"
                                   }, inplace = True
                             )
      
      return date_time
     
New_Cols = Rename_Cols(date_time)
New_Cols
```
