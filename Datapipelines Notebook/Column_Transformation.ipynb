```python
data_cols = data.filter(items=['Campaign_Name', 'Status', 'Payout'])
data_cols.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose  import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
X_train, X_test, y_train, y_test = train_test_split(data_cols.drop(columns=['Payout']), data_cols['Payout'],
                                                    test_size=0.2,  random_state=42)
X_train
```

```python
transfromer = ColumnTransformer(
     [
         ("OneHotEncoding", OneHotEncoder(sparse_output=False, handle_unknown='ignore'), ['Campaign_Name', 'Status']) 
     ], remainder='passthrough'
     )
```

```python
transfromer_output = transfromer.fit_transform(X_train)
transfromer_X_test = transfromer.transform(X_test)
transfromer.set_output(transform='pandas')
```
