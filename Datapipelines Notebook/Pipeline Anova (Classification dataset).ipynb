# Pipeline ANOVA SVM

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Generate classification dataset

X, y = make_classification(n_features= 20, n_informative=3, n_redundant=0,
                           n_classes=2, n_clusters_per_class=2, random_state=42)
```

```python
# Convert to DataFrame
import pandas as pd
X_df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])
y_df = pd.Series(y, name='target')
# Display the DataFrame
X_df['target'] = y_df  # Adding target column to the DataFrame if desired
X_df.head()
```

```python
X_df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_test
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import make_pipeline
from sklearn.svm import LinearSVC
anova_filter = SelectKBest(f_classif, k = 3)
clf = LinearSVC()
anova_svm = make_pipeline(anova_filter, clf)
anova_svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = anova_svm.predict(X_test)
print(classification_report(y_test, y_pred))
```
