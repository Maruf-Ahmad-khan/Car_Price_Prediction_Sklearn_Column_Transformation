```python
from tabulate import tabulate
def Select_Cols(df:pd.DataFrame)->pd.DataFrame:
     
     try:
          
          df_cols = df.filter(items=['Weight', 'Height'])
          return df_cols
     
     except Exception as e:
          print(f"Error occurred: {e}")
          return []
     
df_cols = Select_Cols(df)
print(tabulate(df_cols, headers='keys', tablefmt='grid'))
```

```python
# Find Correlation and Visualization
from tabulate import tabulate
def Correlation(df_cols: pd.DataFrame)->pd.DataFrame:
     
     try:
          
          return df_cols.corr()
     
     except Exception as e:
          print(f'Error Occurred: {e}')
          return []
     
def Scatter(df_cols:pd.DataFrame):
     
     try:
          
          plt.scatter(df_cols['Weight'], df_cols['Height'])
          plt.xlabel("Weight")
          plt.ylabel("Height")
          plt.show()
          
     except Exception as e:
          print(f"Error Occurred: {e}")
          
corr_cols = Correlation(df_cols)
print(tabulate(corr_cols, headers='keys', tablefmt='grid'))
```

```python
from typing import Tuple
from sklearn.preprocessing import StandardScaler
import pandas as pd

def standardize(
    X_train: pd.DataFrame,
    X_test: pd.DataFrame,
    return_df: bool = True
) -> Tuple[pd.DataFrame, pd.DataFrame, StandardScaler]:
    scaler = StandardScaler()
    X_train_arr = scaler.fit_transform(X_train)
    X_test_arr = scaler.transform(X_test)

    if return_df:
        X_train_scaled = pd.DataFrame(X_train_arr, columns=X_train.columns, index=X_train.index)
        X_test_scaled  = pd.DataFrame(X_test_arr,  columns=X_test.columns,  index=X_test.index)
    else:
        X_train_scaled = X_train_arr
        X_test_scaled  = X_test_arr

    return X_train_scaled, X_test_scaled, scaler
```

```python
from typing import Optional, Tuple
from sklearn.linear_model import LinearRegression
import pandas as pd

def train_linear_regression(
    X_train: pd.DataFrame,
    y_train: pd.Series
) -> Tuple[Optional[LinearRegression], Optional[float]]:
    """
    Train a simple LinearRegression and return the fitted model and training R^2.
    Raises ValueError for invalid inputs.
    """
    if X_train is None or y_train is None:
        raise ValueError("X_train and y_train must be provided")
    if len(X_train) == 0:
        raise ValueError("X_train is empty")
    if len(X_train) != len(y_train):
        raise ValueError("X_train and y_train must have the same number of rows")
    # Ensure numeric
    if not all(dtype.kind in "biufc" for dtype in X_train.dtypes):  # bool, int, unsigned, float, complex
        raise ValueError("X_train must contain only numeric columns")

    model = LinearRegression()
    model.fit(X_train, y_train)
    r2 = model.score(X_train, y_train)
    return model, r2

slr, train_r2 = train_linear_regression(X_train, y_train)
print("Train R^2:", train_r2)
print("Coef:", slr.coef_, "Intercept:", slr.intercept_)
```

```python
def predict_linear_regression_plot(X_train:pd.DataFrame, y_train:pd.Series, X_test:pd.DataFrame, y_test:pd.Series):
     
     try:
          
          slr = LinearRegression()
          slr.fit(X_train, y_train)
          y_pred = slr.predict(X_test)
          
          plt.scatter(X_test, y_test, color='blue', label='Actual')
          plt.scatter(X_test, y_pred, color='red', label='Predicted')
          plt.xlabel("Weight")
          plt.ylabel("Height")
          plt.legend()
          plt.show()
          
     except Exception as e:
          print(f"Error Occurred: {e}")
          
predict_linear_regression_plot(X_train, y_train, X_test, y_test)
```

```python
# performance Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
def performance_metrics(y_test:pd.Series, y_pred:pd.Series):
     
     try:
          
          mae = mean_absolute_error(y_test, y_pred)
          mse = mean_squared_error(y_test, y_pred)
          r2 = r2_score(y_test, y_pred)
          
          return mae, mse, r2
     
     except Exception as e:
          print(f"Error Occurred: {e}")
          return []
y_pred = slr.predict(X_test)
mae, mse, r2 = performance_metrics(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# ols from statsmodels.api import OLS, add_constant
import statsmodels.api as sm
def ols_statsmodels(X_train:pd.DataFrame, y_train:pd.Series):
     
     try:
          
          X_train_const = sm.add_constant(X_train)
          ols_model = sm.OLS(y_train, X_train_const).fit()
          return ols_model
     
     except Exception as e:
          print(f"Error Occurred: {e}")
          return []
ols_model = ols_statsmodels(X_train, y_train)
print(ols_model.summary())
```
