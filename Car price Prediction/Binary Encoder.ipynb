```python
import pandas as pd
import category_encoders as ce
import numpy as np
data = {'Item' : ['Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8'],
        'Fruit':['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry', 'Fig', 'Grap', 'Honeydew']}
df = pd.DataFrame(data)
df
```

```python
# Initialize the Binary Encoder
encoder = ce.BinaryEncoder(cols = ['Fruit'], return_df=True)

# Fit and transform the data
df_encoded = encoder.fit_transform(df)
print(df_encoded)
```

```python
def Binary_Encoder(df):
     encoder = ce.BinaryEncoder(cols = ['Fruit'], return_df= True)
     return encoder.fit_transform(df)


data = {'Item' : ['Item1', 'Item2', 'Item3', 'Item4', 'Item5', 'Item6', 'Item7', 'Item8'],
        'Fruit':['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry', 'Fig', 'Grap', 'Honeydew']}
df = pd.DataFrame(data)

Binary_Encoder(df)

```
